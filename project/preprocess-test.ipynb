```python
import pandas as pd
```

```python
nlp = stanza.Pipeline(lang='ru', processors='tokenize,lemma,pos')
```

```python
def preprocess(text, nlp=nlp):
    list_of_results = []
    doc = nlp(text)
    for sent in doc.sentences:
        for word in sent.words:
            list_of_results.append([sent.text, word.text, word.lemma, word.start_char, word.end_char])            
    return list_of_results
```

```python
reviews = pd.read_csv('dev_reviews.txt', sep='\t', header=None)
```

```python
reviews
```

```python
ids = reviews[0].values
text_reviews = reviews[1].values
ids_reviews = list(zip(ids, text_reviews))
```

```python
from tqdm import tqdm
import numpy as np
```

```python
result = []
for i_r in tqdm(ids_reviews):
    pre_res = preprocess(i_r[1])
    for res in pre_res:
        to_app = list(i_r)
        to_app.extend(res)
        result.append(to_app)
```
